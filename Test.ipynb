```python
import numpy as np
import matplotlib.pyplot as plt

from uncertainties import ufloat

# Gegebene Daten
x_values = np.array([0.23, 0.43, 0.63, 0.83, 0.13, 0.33, 0.53, 0.73])
y_values = np.array([2.833, 3.861, 4.665, 5.353, 2.156, 3.442, 4.336, 5.094])

# Quadrierte x-Werte berechnen
x_squared = np.square(x_values)

# Plot der quadrierten x-Werte gegen die y-Werte
plt.scatter(x_squared, y_values, label='Beschleunigte Bewegung', color='green', marker='o')

# Achsentitel und Legende hinzufügen
plt.xlabel('t²')
plt.ylabel('s')
plt.legend()

# Plot anzeigen
plt.show()

from sympy import symbols, Eq, solve

# Definition der Symbole
m_W, mu, m_Z, g, a1, a2 = symbols('m_W mu m_Z g a1 a2')

# Gleichung 1
eq1 = Eq(2 * g * (m_W + 0.5) * a1 - m_W * g * mu, 0)

# Gleichung 2
eq2 = Eq(2 * g * (m_W + m_Z + 0.5) * a2 - (m_W + m_Z) * g * mu, 0)



# Lösen des Gleichungssystems nach m_W und mu
solution = solve((eq1, eq2), (m_W, mu))

# Zeige die Lösungen an
print("Lösungen:", solution)
```
